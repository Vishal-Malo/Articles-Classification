```python
import numpy as np
import pandas as pd
from collections import Counter
```

```python
data=pd.read_csv('Articles.csv',encoding="ISO-8859-1")
```

```python
data.head()
```

```python
data['Date'].value_counts()
```

```python
data.drop(columns=['Date'],inplace=True)
```

```python
data.shape
```

```python
article_words=[]
for article in data['Article']:
    article_words+=article.split(" ")
```

```python
len(article_words)
```

```python
for i in range(len(article_words)):
    if not article_words[i].isalpha():
        article_words[i]=""
```

```python
article_words
```

```python
words_dict=Counter(article_words)
words_dict
```

```python
del words_dict['']
```

```python
len(words_dict)
```

```python
words_dict=words_dict.most_common(3000)
```

```python
words_dict
```

```python
features=[]
for article in data["Article"]:
    blob=article.split(" ")
    data1=[]
    for i in words_dict:
        data1.append(blob.count(i[0]))
    features.append(data1)
```

```python
heading_words=[]
for heading in data['Heading']:
    heading_words+=heading.split(" ")
```

```python
for i in range(len(heading_words)):
    if not heading_words[i].isalpha():
        heading_words[i]=""
```

```python
words_dict1=Counter(heading_words)
```

```python
del words_dict1['']
```

```python
words_dict1=words_dict1.most_common(3000)
```

```python
features1=[]
for heading in data["Heading"]:
    blob=heading.split(" ")
    data2=[]
    for i in words_dict1:
        data2.append(blob.count(i[0]))
    features1.append(data2)
```

```python
data['NewsType'].value_counts()
```

```python
data['NewsType'].replace({'sports':1,'business':0}, inplace=True)
```

```python
article_final=np.array(features)
heading_final=np.array(features1)
```

```python
X=np.hstack((article_final,heading_final))
```

```python
y=np.array(data['NewsType'])
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.2,random_state=7)
```

```python
from sklearn.naive_bayes import MultinomialNB
classifier=MultinomialNB()
```

```python
classifier.fit(X_train,y_train)
```
